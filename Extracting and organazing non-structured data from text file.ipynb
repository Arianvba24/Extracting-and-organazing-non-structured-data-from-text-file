```python
# USDT_USD
#The two following scripts extract data from a non-structured text file extracted throught
#web scrapping
file_adress = r"C:\Users\Cash\Documents\pruebas_python\proyectos\USDT_USD.txt"
import pandas as pd





user = []
transactions = []
payment_type = []
comment = []
nueva_oferta = []
price = []
discount = []
last_connection = []
min_amount = []
max_amount = []
availability = []





x = 0

counter = 1
with open(file_adress, encoding='utf-8') as file:
    for line in file:
        value = line.strip()
        if counter == 1:
            user.append(value)
            counter = counter + 1
        elif counter == 2:
            transactions.append(value)
            counter = counter + 1
        elif counter == 3:
            payment_type.append(value)
            counter = counter + 1
        elif counter == 4 and (value == "Nueva oferta" or value[0].isdigit()):
            nueva_oferta.append(value)
            comment.append("None comment")
            counter = counter + 2
        elif counter == 4 and (value != "Nueva oferta" or value[0].isdigit() == False):
            comment.append(value)
            counter = counter + 1
            x = 1
        elif counter == 5 and x == 1:
            nueva_oferta.append(value)
            counter = counter + 1
            x = 0
        elif counter == 6:
            price.append(value)
            counter = counter + 1
        elif counter == 7 and "%" in value:
            discount.append(value)
            counter = counter +1
        elif counter == 7 and "%" not in value:
            discount.append("No hay descuento")
            last_connection.append(value)
            counter = counter + 2
        elif counter == 8:
            last_connection.append(value)
            counter = counter + 1
        elif counter == 9 and "Min" not in value:
            availability.append(value)
            min_amount.append("No hay")
            max_amount.append("No hay")
            counter = 1
            continue
        elif counter == 9 and "Min" in value:
            min_amount.append(value)
            counter = counter + 1
        elif counter == 10 and "Max" in value:
            max_amount.append(value)
            counter = 1
        
        
values = {
    "User" : user,
    "Transactions" : transactions,
    "Payment_type" : payment_type,
    "Comment" : comment,
    "New offer" : nueva_oferta,
    "Price" : price,
    "Discount" : discount,
    "Last Connection" : last_connection,
    "Min amount" : min_amount,
    "Max amount" : max_amount,
#     "Availability" : availability
    

    
}


df = pd.DataFrame(values)       

df
            
        
            

```

```python
# USDT_USD

lista = list(range(1,df.shape[0]+1))

df["Indice"] = lista


df.set_index("Indice", inplace = True)
df["Currency"] = df["Price"].apply(lambda x: x[-4:])
df["Price"] = df["Price"].str.replace(" USD","")
df["Min amount"] = df["Min amount"].str.replace("Min sale: ", "")
df["Min amount"] = df["Min amount"].str.replace(" USD","")
df["Max amount"] = df["Max amount"].str.replace("Max sale: ", "")
df["Max amount"] = df["Max amount"].str.replace(" USD","")

df.to_excel(r"C:\Users\Cash\Documents\pruebas_python\proyectos\paxful_USDT_USD.xlsx", index = False)
```
